```python
import numpy as np
import matplotlib.pyplot as plt

# Generate noisy signal
np.random.seed(0)
t = np.linspace(0, 1, 500, endpoint=False)
clean_signal = np.sin(2 * np.pi * 5 * t)
noise = np.random.normal(0, 0.5, clean_signal.shape)
noisy_signal = clean_signal + noise

# Define Gradient Descent function
def gradient_descent(noisy_signal, learning_rate=0.01, n_iterations=100):
    filtered_signal = np.copy(noisy_signal)
    for _ in range(n_iterations):
        gradient = np.gradient(filtered_signal)
        filtered_signal -= learning_rate * gradient  # Update using gradient
    return filtered_signal

# Apply gradient descent for filtering
filtered_signal = gradient_descent(noisy_signal)

# Plot results
plt.figure(figsize=(20, 8))

plt.subplot(1,3,1)
plt.plot(t, noisy_signal, label='Noisy Signal', alpha=0.6)
plt.legend()
plt.title("noisy signal using Gradient Descent")
plt.show()



plt.figure(figsize=(20, 8))
plt.subplot(1,3,2)
plt.plot(t, clean_signal, label='Clean Signal', linestyle='dashed')
plt.legend()
plt.title("clean signal using Gradient Descent")
plt.show()


plt.figure(figsize=(20, 8))
plt.subplot(1,3,3)
plt.plot(t, filtered_signal, label='Filtered Signal', linewidth=2)
plt.legend()
plt.title("Signal Filtering using Gradient Descent")
plt.show()
```

```python
def gradient_descent(func, grad_func, x_init, learning_rate=0.1, n_iterations=100):
    """Perform gradient descent to minimize a function."""
    x = x_init
    trajectory = [x]  # To visualize the path
    for _ in range(n_iterations):
        grad = grad_func(x)
        x -= learning_rate * grad
        trajectory.append(x)
    return x, trajectory

# Example: Minimize f(x) = (x - 2)^2
def func(x):
    return (x - 2) ** 2

def grad_func(x):
    return 2 * (x - 2)

x_init = 10.0  # Starting point
x_min, trajectory = gradient_descent(func, grad_func, x_init)

# Results
print("Minimum at x =", x_min)
x_vals = np.linspace(-5, 15, 100)
y_vals = func(x_vals)

# Plotting
plt.plot(x_vals, y_vals, label="f(x) = (x - 2)^2")
plt.scatter(trajectory, [func(x) for x in trajectory], color='red', label="Steps")
plt.title("Gradient Descent Optimization")
plt.legend()
plt.show()
```
